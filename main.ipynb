单个课程课评长度直方图

```python
url = "https://icourse.club/course/21642/"

comments = web.GetComment(url)

# 评课长度分布直方图
LengthDistribution = list(map(len,comments))
sns.displot(LengthDistribution)
plt.show()

```

```python
def TheWordCloud(words):
#绘制云图
    bg="./USTC.jpg"
    imgbg=imread(bg)

    a=WordCloud(font_path="./font/simkai.ttf", background_color='white', width=1000, height=800,mask=imgbg,colormap="summer")
    
    a.generate(words)
    plt.imshow(a)
    plt.axis("off")
    plt.show()


web.TheWordCloud(comments)
```
